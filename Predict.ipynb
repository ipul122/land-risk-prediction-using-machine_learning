```python
import pandas as pd
import joblib
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

df = pd.read_csv("data_prediksi.csv")
X = df.drop(columns=["y_pred"])
y = df["y_pred"]
```

```python
X.head()
```

```python
len(X)
```

## LOAD

```python
sc = joblib.load("scaler.pkl")
best_model_XGB = joblib.load("best_model_XGB.pkl")
best_model_RF  = joblib.load("best_model_RF.pkl")
```

## PRED

```python
X_scaled = sc.transform(X)

y_pred_XGB = best_model_XGB.predict(X_scaled)
y_pred_RF  = best_model_RF.predict(X_scaled)
```

## EVAL

```python
print("XGB Accuracy:", accuracy_score(y, y_pred_XGB))
print(classification_report(y, y_pred_XGB))
print(confusion_matrix(y, y_pred_XGB))

print("\nRF Accuracy:", accuracy_score(y, y_pred_RF))
print(classification_report(y, y_pred_RF))
print(confusion_matrix(y, y_pred_RF))
```
